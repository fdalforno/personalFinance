```python
import numpy as np
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt

import ipywidgets as widgets
from IPython.display import display

from randomgenerator import generate_price_path, BULL, SIDEWAYS, BEAR
from trading import compare_strategies_with_costs
```

Confrontiamo una strategia buy and hold rispetto ad una strategia basata su segnali per fare questo useremo un generatore random che simula l'andamento di un mercato. 

```python
aggregated = pd.DataFrame()


years_slider = widgets.IntSlider(
    value=10,
    min=1,
    max=30,
    step=1,
    description='Anni:',
    continuous_update=False
)
    
simulations_slider = widgets.IntSlider(
    value=1000,
    min=100,
    max=10000,
    step=100,
    description='Simulazioni:',
    continuous_update=False
)
    
regime_dropdown = widgets.Dropdown(
    options=[('Bull', BULL), ('Sideways', SIDEWAYS), ('Bear', BEAR)],
    value=BULL,
    description='Regime:',
)
    
start_price_input = widgets.FloatText(
    value=100.0,
    description='Prezzo iniziale:',
    step=10.0
)



display(years_slider, simulations_slider,regime_dropdown,start_price_input)
```

Simuliamo ora le nostre strategie inserendo:

* costi di transazione pari allo 0.1%
* slippage, questo fattore è dovuto al fatto che al momento dell'ordine e la sua esecuzione passa del tempo e il prezzo cambia. lo impostiamo al 0.05%
* tassazione ogni qual volta creiamo un guadagno questo viene tassato in Italia la tassazione sulle azioni è pari al 26%

Con un po di pazienza vediamo il risultato delle varie strategie

```python
price_paths = generate_price_path(
    start_price=start_price_input.value,
    start_regime=regime_dropdown.value,
    num_years=years_slider.value,
    num_simulations=simulations_slider.value
)

compare_strategies_with_costs(price_paths)
```
