## Il Gioco delle Figurine 
Immagina che tu e i tuoi amici usiate le figurine dei calciatori come moneta per scambiarvi le merende a scuola.
Oggi, con 1 figurina, puoi avere un pacchetto di cracker.
Immagina però che un giorno arrivi un camion e regali a ogni bambino della scuola 1000 figurine. All'improvviso, tutti hanno le tasche piene!
Cosa succederà il giorno dopo? Se vuoi i cracker del tuo amico, lui non vorrà più solo 1 figurina (ne ha già tantissime!). 
Ti dirà: "Per i miei cracker ora ne voglio 10".
L'inflazione è esattamente questo: non è la merenda che è diventata magica o più buona, è la tua figurina che è diventata "meno rara" e quindi vale meno. Quando nel mondo ci sono troppi soldi o le cose da comprare diventano difficili da produrre, i prezzi salgono.

Ora ho semplificato moltissimo il ragionamento ma la domanda è perchè non possiamo eliminarla?

Ecco i due motivi principali per cui un po' di inflazione ci serve:
* A. La Paura del "Prezzo che Scende" (Deflazione)
Se sapessi che domani un giocattolo costerà meno di oggi, e dopodomani ancora meno, cosa faresti? Aspetteresti per sempre a comprarlo.
Se tutti smettono di comprare in attesa che i prezzi scendano, i negozi chiudono, le fabbriche non producono più nulla e i papà e le mamme perdono il lavoro. Un pochino di inflazione (i prezzi che salgono un pochino) ci spinge a far girare l'economia oggi.

* B. L'economia che cresce
Immagina che l'economia sia un bambino che cresce. Se il bambino diventa più grande, ha bisogno di vestiti più grandi. Se l'economia produce più cose e ci sono più persone, servono "più figurine" in giro. L'inflazione è spesso l'effetto collaterale della crescita: è il segno che la gente ha voglia di comprare e le aziende di produrre.

Questo calcolo però ci aiuta a introdurre un nuovo concetto che si chiama attualizzazione se volessimo sapere quanto dovremmo avere oggi per avere un controvalore pari a 100€ quanti soldi dovremmo avere in cassa?

Questo introduce questa formula:

$$
    VA = \frac{VF}{(1+r)^n}
$$

Vediamo come leggere questa formula VA sta per valore attuale, VF sta per valore futuro r rappresenta il tasso di interesse e n il numero degli anni che abbiamo davanti.

```python
future_value = 100
r = -0.03
n = 20.0

round(future_value / (1+r)**n,2)
```

Facciamo il nostro esempio se fra 20 anni volessimo avere i nostri 100€ oggi dovremmo avere in cassa un importo pari a 184€.

Ora assumendo che per l'università di nostro figlio serviranno 10000€ l'anno per 5 anni quali saranno i nostri flussi di cassa futuri?
Dovremmo andare sommare tutti i flussi di cassa.

Il valore $V(t_i)$ rappresenta il debito di 1€ al tempo $t_i$ se assumiamo che non ci saranno cambiamenti avremo questo risultato

$$
V(t_i) = \frac{1}{(1+r)^t}
$$

Perciò dovremo pensare di risparmiare questo valore

$$
PV(L) = \sum_{i=1}^{k} V(t_i) L_i
$$


```python
def discount(t, r):
    """
    Compute the price of a pure discount bond that pays $1 at time t where t is in years and r is the annual interest rate
    """
    return (1+r)**(-t)
```

```python
b = discount(19, -.03)
b
```

```python
def pv(l, r):
    """
    Compute the present value of a list of liabilities given by the time (as an index) and amounts
    """
    dates = l.index
    discounts = discount(dates, r)
    return (discounts*l).sum()
```

```python
liabilities = pd.Series(data=[10, 10, 10, 10, 10], index=[19, 20, 21, 22, 23])
liabilities
```

Il problema è che l'inflazione cambia anno per anno e non è mai fissa attorno ad un valore quello che dobbiamo fare è provare a simulare l'andamento attraverso una simulazione.

Partiamo con un modellazione semplice ma che già ci permette di estrarre qualche info utile useremo la tecnica del "bootstrapping" potremmo tradurla in italiano con "facciamo con quello che abbiamo" praticamente useremo i dati che abbiamo e con un resampling li proietteremo nel futuro prenderemo come riferimento gli ultimi 30 anni.

```python
def naive_inflaction(n_years: int, n_scenarios: int, n_past_years: int, ):
    data = infl_stor[-n_past_years:]
    infl_simulate = np.random.choice(data,size=(n_scenarios,n_years),replace=True)
    infl_cum = np.prod(1 + infl_simulate,axis=1)

    median = np.median(infl_cum)
    lower = np.percentile(infl_cum,5)
    upper = np.percentile(infl_cum,95)

    inf_mean = (median) ** (1/n_years) - 1
    inf_upper = (upper) ** (1/n_years) - 1

    plt.figure(figsize=(10, 6))
    plt.hist(infl_cum, bins=60, color='skyblue', edgecolor='navy', alpha=0.7)
    plt.axvline(median, color='green', linestyle='--', linewidth=2,
            label=f'Mediana inflazione {inf_mean * 100:,.2f}%')
    plt.axvline(upper, color='red', linestyle='--', linewidth=2,
            label=f'Caso pessimo {inf_upper * 100:,.2f}%')

    plt.title(f"Valore medio di inflazione da usare su {n_years} anni pari a {inf_mean * 100:,.2f}%")



    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
```

```python
import ipywidgets as widgets
from IPython.display import display

sim_controls = widgets.interactive(naive_inflaction, 
                                   n_years=widgets.IntSlider(min=19, max=24, step=1, value=19),
                                   n_scenarios=widgets.IntSlider(min=1000, max=10000, step=5, value=5000),
                                   n_past_years=widgets.IntSlider(min=20, max=len(anni_storici), step=1, value=30)
                                   )
display(sim_controls)
```
